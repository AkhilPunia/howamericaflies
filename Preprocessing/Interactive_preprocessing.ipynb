```python
import pandas as pd
import numpy as np

```

```python
df1 = pd.read_csv("D:/Columbia/Fall 2018/EDAV/Ameya Work final/Airline_Data_v31/01_year_data.csv")
```

```python

list_city = [
"New York, NY" ,
"Atlanta, GA" ,
"Chicago, IL" ,
"Dallas, TX" ,
"Denver, CO" ,
"Los Angeles, CA" ,
"Charlotte, NC" ,
"Phoenix, AZ" ,
"San Francisco, CA",
"Seattle, WA" 
]

total = []
for city in list_city:
    total.append(df1[df1["OriginCityName"] ==  city])
result = pd.concat(total)
total = []
for city in list_city:
    total.append(result[result["DestCityName"] ==  city])
main_result = pd.concat(total)
```

```python
k = ["Month", "Reporting_Airline", "Origin", "OriginCityName", "Dest", "DestCityName", "ArrDelay"]
month_list = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
main_result = main_result[k]
dictionary = {}
```

```python
dictionary = {}
for s_city in list_city:
    dictionary[s_city] = {}
    print "city " + s_city
    for d_city in list_city:
        if s_city == d_city:
            continue;
        dictionary[s_city][d_city] = {}
        for month in range(1, 13):
            dictionary[s_city][d_city][month] = {}
            temp_res1 = main_result[(main_result["OriginCityName"] == s_city) 
                                    & (main_result["DestCityName"] == d_city)
                                    & (main_result["Month"] == month)]
            s_airport_list = list(set(temp_res1["Origin"]))
            for s_airport in s_airport_list:
                dictionary[s_city][d_city][month][s_airport] = {}
                temp_res2 = temp_res1[(temp_res1["Origin"] == s_airport)]
                d_airport_list = list(set(temp_res2["Dest"]))
                for d_airport in d_airport_list:
                    dictionary[s_city][d_city][month][s_airport][d_airport] = {}
                    temp_res3 = temp_res2[(temp_res2["Dest"] == d_airport)]
                    airline_list = list(set(temp_res3["Reporting_Airline"]))
                    for airline in airline_list:
                        temp_res4 = temp_res3[temp_res3["Reporting_Airline"] == airline]
                        nflights = len(temp_res4)
                        if nflights == 0 :
                            break;
                        delay = 0 
                        for i in range(nflights):
                            this_delay = temp_res4.iloc[i]["ArrDelay"]
                            if (np.isnan(this_delay)):
                                this_delay = 0
                            delay = delay + this_delay
                        dictionary[s_city][d_city][month][s_airport][d_airport][airline] = (delay, nflights)
   
                
                
            
            
```
